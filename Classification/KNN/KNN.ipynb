# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing Data

```python
data=pd.read_csv(r"C:\Users\yaksh\OneDrive\Desktop\programing\ML\ML data\Dataset_master.xlsx - Loan Distribution KNN Classific.csv")
data.head()
```

```python
X=data.iloc[:,0:2].values
Y=data.iloc[:,-1:].values
```

```python
print(X)
```

```python
print(Y)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
print(X)
```

# Spliting Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python

print(Y_train)
```

```python
print(Y_test)
```

# ML Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=6,metric="minkowski",p=2)
classifier.fit(X_train,Y_train)
```

# Prediction

```python
print(classifier.predict(sc.fit_transform([[1,745]])))
```

```python
a=classifier.predict(X_test)
a=a.reshape(len(a),1)
b=Y_test
b=b.reshape(len(b),1)
print(np.concatenate((a,b),1))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(b,a))
cm=confusion_matrix(b,a)
cm
```

```python
import seaborn as sn
sn.heatmap(cm,annot=True)
plt.xlabel('True')
plt.ylabel('predicted')
```
