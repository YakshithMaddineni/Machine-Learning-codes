# importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# importing dataset

```python
data_set=pd.read_csv(r"C:\Users\yaksh\OneDrive\Desktop\programing\ML\ML data\Dataset_master.xlsx - Instagram Ads Logistic Regressi.csv")
data_set.head()
```

```python
X=data_set.iloc[:,0:2].values
Y=data_set.iloc[:,-1:].values
```

```python
print(X)
```

```python
print(Y)
```

# Test train Spliting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()                             # we use feature scaling because we will make the values between -3 and +3
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

# Machine learning

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
classifier.predict(sc.transform([[25,45000]]))
```

# Predicting Values

```python
y_pred = classifier.predict(X_test)
np.concatenate((y_pred.reshape(len(y_pred),1),Y_test.reshape(len(Y_test),1)),axis=1)
```

# Accuracy score

```python
from sklearn.metrics import accuracy_score,r2_score
print(accuracy_score(y_pred,Y_test))
r2_score(y_pred,Y_test)
```
