```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\yaksh\OneDrive\Desktop\programing\ML\ML data\Dataset_master.xlsx - Salary Data Support Vector regr.csv")
data.head()
X=data.iloc[:,1:2].values
Y=data.iloc[:,-1:].values
print(X)
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler
X_scale = StandardScaler()
Y_scale = StandardScaler()
X=X_scale.fit_transform(X)
Y=Y_scale.fit_transform(Y)
print(X)
print(Y)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg_model = LinearRegression()
lin_reg_model.fit(X,Y)
```

```python
y_lin=lin_reg_model.predict(X)
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,y_lin,color='blue')
plt.title('EXPERIENCE VS SALARY')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=6)
x_poly=poly.fit_transform(X)
poly_reg=LinearRegression()
poly_reg.fit(x_poly,Y)
```

```python
y_poly=poly_reg.predict(x_poly)
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,y_poly,color='blue')
plt.title('EXPERIENCE VS SALARY')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
from sklearn.metrics import r2_score
r2_score(Y,y_poly)
```

```python
from sklearn.metrics import r2_score
r2_score(Y,y_lin)
```
